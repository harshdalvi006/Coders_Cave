# **Information about the each certificates**

1. "U" = Universal. Used across several systems

2. "A" = In the Indian system, this is for 'Adults Only'

3. "UA" and "U/A" = Indian equivalent of UK PG. i.e. parental guidance recommended

4. "R" = Restricted. In the USA, under 17 requires accompanying parent or adult guardian

5. "PG-13" = Parents strongly cautioned – Some material may be inappropriate for children under 13

6. "PG" = Parental guidance suggested – Some material may not be suitable for children

7. "Passed" = Pre-dates cinema classification. There is no known rating, but was allowed to be released

8. "G" = General audiences – All ages admitted. Gonna fold into "U"

9. "Approved" = Used interchangably with "Passed" as far as I can tell 10. "TV-PG" = Televsion movie rated PG. Fold into "PG" 11. "GP" = All ages admitted – Parental guidance suggested. Fold into "PG" 12. "TV-14" = Fold into "PG-13" 13. "16" = Irish classification. Fold into "R" 14. "TV-MA" = Fold into "R" 15. "Unrated" = Unrated. I am going to assume this is the same as the NaNs

```python
# Let's bring some categories together by using .loc() method
df.loc[df['Certificate']=='U/A','Certificate'] = 'UA'
df.loc[df['Certificate']=='G','Certificate'] = 'U'
df.loc[df['Certificate']=='Approved', 'Certificate'] = 'Passed'
df.loc[df['Certificate']=='TV-PG', 'Certificate'] = 'PG'
df.loc[df['Certificate']=='GP', 'Certificate'] = 'PG'
df.loc[df['Certificate']=='TV-14', 'Certificate'] = 'PG-13'
df.loc[df['Certificate']=='16', 'Certificate'] = 'R'
df.loc[df['Certificate']=='TV-MA', 'Certificate'] = 'R'

# assuming missing certicate data means movies are unrated
# filling null values with Unrated one
df.loc[df['Certificate'].isna(), 'Certificate'] = 'Unrated'

# Counting values of Certificate Column
df['Certificate'].value_counts()
```

```python
# Plotting Percentage Distribution of movies on the basis of 'Certificate'
plt.figure(figsize=(12,10))
df['Certificate'].value_counts().plot(kind='pie',autopct='%.2f%%')
plt.title('Certificate Count Percentage Distribution')
plt.legend(df['Certificate'].value_counts().index)
plt.show()
```

In this plot, The most movies which are of certificate 'U' are rated in top 1000 of imdb because the U certificate movies are friendly to all categories of the people(adults,old,children) etc.
